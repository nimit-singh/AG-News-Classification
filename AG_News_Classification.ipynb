# 2. Text Preprocessing
* Text data is cleaned and prepared using the following steps:
    * Convert text to lowercase
    * Remove punctuation
    * Remove stopwords
    * Tokenize text into words

These steps help improve the performance of machine learning models.

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)
    
    return " ".join(filtered_tokens) 
```

```python
df_train['Preprocess_Title'] = df_train['Title'].apply(preprocess)
```

```python
df_train['Preprocess_Description'] = df_train['Description'].apply(preprocess)
```

```python
df_train.head()
```

```python
df_train['combined_text'] = df_train['Preprocess_Title'] + " " + df_train['Preprocess_Description']
```

```python
df_train.head()
```

# 3. Feature Extraction
* The cleaned text data is converted into numerical form using TF-IDF Vectorization, which measures the importance of words in the dataset.

```python
X = df_train['combined_text']
y = df_train['Class Index']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# 4. Model Training

Machine learning algorithms are used to train the classification model.

Algorithms used:

* Logistic Regression

* KNN Model

* Naive Bayes

* Decision Tree Classifier

* Linear Support Vector Machine

These algorithms learn patterns from the dataset and classify news articles into their respective categories.


# Logistic Regression

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf_LR = Pipeline([
    ("vectorizer",TfidfVectorizer()),
    ("Logistic Regression",LogisticRegression(max_iter=2000))
])
```

```python
clf_LR.fit(X_train,y_train)
```

```python
y_pred_LR = clf_LR.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred_LR))
```

```python
new_sentence = ["Apple launches new AI powered iPhone"]

prediction = clf_LR.predict(new_sentence)

print(prediction)
```

# KNN Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline

clf_KNN = Pipeline([
    ("vectorizer",TfidfVectorizer()),
    ("KNN",KNeighborsClassifier())
])
```

```python
clf_KNN.fit(X_train,y_train)
```

```python
y_pred_KNN = clf_KNN.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_KNN))
```

```python
new_sentence = ["Apple launches new AI powered iPhone"]

prediction = clf_KNN.predict(new_sentence)

print(prediction)
```

# Multinomial Naive Bayes

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

clf_NB = Pipeline([
    ("vectorizer",TfidfVectorizer()),
    ("NB",MultinomialNB())
])
```

```python
clf_NB.fit(X_train,y_train)
```

```python
y_pred_NB = clf_NB.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_NB))
```

# Decision tree classifier

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier 
from sklearn.pipeline import Pipeline

clf_tree = Pipeline([
    ("vectorizer",TfidfVectorizer()),
    ("KNN",DecisionTreeClassifier())
])
```

```python
clf_tree.fit(X_train,y_train)
```

```python
y_pred_tree = clf_tree.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_tree))
```

# Support vector machine Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC 
from sklearn.pipeline import Pipeline

clf_SVC = Pipeline([
    ("vectorizer",TfidfVectorizer()),
    ("SVC",LinearSVC())
])
```

```python
y_pred_svc = clf_SVC.predict(X_test)
```
